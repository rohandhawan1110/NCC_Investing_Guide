## Query 1.4
In a separate array , calculate the monthwise VWAP (Volume Weighted Average Price ) of the stock. 
( VWAP = sum(price*volume)/sum(volume) ) 
To know more about VWAP , visit - VWAP definition

Similarly, create a column ‘Year’ and initialize it. Then use the 'groupby()' function by month and year. Finally, calculate the vwap value for each month (i.e. for each group created).

```python
df["Month"]=df["Date"].dt.month
df["Year"]=df["Date"].dt.year
df["VWAP"]=df["Open Price"]+df["High Price"]+df["Close Price"]
df["VWAP"]=df["VWAP"]/3
df["VWAP"]=df["VWAP"]*df["Total Traded Quantity"]
df1=df[["Month","Year","Total Traded Quantity","VWAP"]].groupby(["Month","Year"]).sum()
df1["VWAP"]=df1["VWAP"]/df1["Total Traded Quantity"]

df1
```

## Query 1.5
Write a function to calculate the average price over the last N days of the stock price data where N is a user defined parameter. Write a second function to calculate the profit/loss percentage over the last N days.
Calculate the average price AND the profit/loss percentages over the course of last -
1 week, 2 weeks, 1 month, 3 months, 6 months and 1 year.
{Note : Profit/Loss percentage between N days is the percentage change between the closing prices of the 2 days }

```python
def avg(N):
    return ((df["Close Price"].tail(N).sum())/N)

print("Average Price for last")
print("1 week: ",avg(5))
print("2 weeks: ",avg(10))
print("1 month: ",avg(21))
print("3 months: ",avg(63))
print("6 months: ",avg(127))
print("1 year: ",avg(253))
```

```python
def plp(N):
    diff=(df["Close Price"].tail(1).sum())-(df["Close Price"].tail(N).sum()-df["Close Price"].tail(N-1).sum())
    if diff==0:
        return 0
    elif diff<0:
        loss=(diff/df["Close Price"].tail(1).sum())*100
        return loss
    else:
        profit=(diff/df["Close Price"].tail(1).sum())*100
        return profit

print("Profit/Loss percentage for last")
print("1 week: ",plp(5))
print("2 weeks: ",plp(10))
print("1 month: ",plp(21))
print("3 months: ",plp(63))
print("6 months: ",plp(127))
print("1 year: ",plp(253))
```

## Query 1.6
 Add a column 'Day_Perc_Change' where the values are the daily change in percentages i.e. the percentage change between 2 consecutive day's closing prices. Instead of using the basic mathematical formula for computing the same, use 'pct_change()' function provided by Pandas for dataframes. You will note that the first entry of the column will have a ‘Nan’ value. Why does this happen? Either remove the first row, or set the entry to 0 before proceeding.

```python
df["Day_Perc_Change"]=df["Close Price"].pct_change()*100
df.dropna(axis=0, how="any", inplace=True)
```

```python
df.reset_index(inplace=True)
df.head()
```

## Query 1.7
Add another column 'Trend' whose values are:
'Slight or No change' for 'Day_Perc_Change' in between -0.5 and 0.5
'Slight positive' for 'Day_Perc_Change' in between 0.5 and 1
'Slight negative' for 'Day_Perc_Change' in between -0.5 and -1
'Positive' for 'Day_Perc_Change' in between 1 and 3
'Negative' for 'Day_Perc_Change' in between -1 and -3
'Among top gainers' for 'Day_Perc_Change' in between 3 and 7
'Among top losers' for 'Day_Perc_Change' in between -3 and -7
'Bull run' for 'Day_Perc_Change' >7
'Bear drop' for 'Day_Perc_Change' <-7

```python
df['Trend'] = np.nan
for i in range(df.shape[0]):
    if (df["Day_Perc_Change"][i]>=-0.5 and df["Day_Perc_Change"][i]<0.5):
        df["Trend"][i]="Slight or No change"
    if (df["Day_Perc_Change"][i]>=0.5 and df["Day_Perc_Change"][i]<1):
        df["Trend"][i]="Slight positive"
    if (df["Day_Perc_Change"][i]>=-1 and df["Day_Perc_Change"][i]<-0.5):
        df["Trend"][i]="Slight negative"
    if (df["Day_Perc_Change"][i]>=1 and df["Day_Perc_Change"][i]<3):
        df["Trend"][i]="Positive"
    if (df["Day_Perc_Change"][i]>=-3 and df["Day_Perc_Change"][i]<-1):
        df["Trend"][i]="Negative"
    if (df["Day_Perc_Change"][i]>=3 and df["Day_Perc_Change"][i]<7):
        df["Trend"][i]="Among top gainers"
    if (df["Day_Perc_Change"][i]>=-7 and df["Day_Perc_Change"][i]<-3):
        df["Trend"][i]="Among top losers"
    if (df["Day_Perc_Change"][i]>=7):
        df["Trend"][i]="Bull run"
    if (df["Day_Perc_Change"][i]<-7):
        df["Trend"][i]="Bear drop"
df.head()
```
